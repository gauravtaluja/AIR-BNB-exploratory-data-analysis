### count of host_id

```python
df['host_id'] = df['host_id'].astype(str)  # converting hostid from int into string
```

```python
hostid = df.host_id.value_counts()[:10]
x = hostid.index
y = hostid.values
X = []
Y = []
for i in x:
    X.append(i)
for i in y:
    Y.append(i)
plt.bar(X,Y,color=['black', 'red', 'green', 'blue', 'cyan'])
plt.xticks(rotation  = 90)
plt.show()
```

### PEARSON'S CORRELATION
#The Pearson correlation coefficient (named for Karl Pearson) can be used
#to summarize the strength of the linear relationship between two data samples.

#The Pearsons correlation coefficient is calculated as the covariance of the two 
#variables divided by the product of the standard deviation of each data sample. It 
#is the normalization of the covariance between the two variables to give an interpretable score.

```python
corr = df.corr(method='pearson')
plt.figure(figsize=(15,8))
sns.heatmap(corr, annot=True)
df.columns
```
